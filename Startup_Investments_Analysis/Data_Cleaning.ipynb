## **Data Cleaning**

### **Column Naming and Selecting**

```python
vc_df.columns
```

```python
vc_df.rename(columns = {' market ':'market', ' funding_total_usd ': 'funding_total_usd'}, inplace = True)
```

```python
 clean_df = vc_df[['name', 'market','country_code' , 'funding_total_usd', 'status', 'state_code', 'funding_rounds', 'founded_at', 'founded_year', 'first_funding_at', 'last_funding_at','seed', 'venture', 'angel', 'private_equity', 'debt_financing', 'round_A',	'round_B',	'round_C' ]]
 clean_df.head(10)
```

```python
clean_df
```

### **All Nan rows delete**

```python
clean_df.isna().sum()
```

```python
na_rows = clean_df[clean_df.isnull().all(axis=1)]
na_rows
```

```python
no_na_df =clean_df.dropna(how='all')
no_na_df
```

### **Data Normalization**

```python
 no_na_df.country_code.unique()
```

```python
no_na_df.funding_total_usd.isna().any()
```

```python
no_na_df[no_na_df['funding_total_usd'] == ' -   ']
```

```python
no_na_df.funding_total_usd.replace(' -   ', np.nan, inplace=True)
```

```python
no_na_df["funding_total_usd"].isna().sum()
```

```python
no_na_df["funding_total_usd"] = no_na_df["funding_total_usd"].str.replace(' ','')
```

```python
no_na_df["funding_total_usd"] = no_na_df["funding_total_usd"].str.replace(',','')
```

```python
no_na_df.head(10)
```

```python
no_na_df['founded_year'] = no_na_df['founded_year'].astype('Int64')
```

```python

no_na_df.founded_year.unique()
```

```python
no_na_df.head(10)
```

```python
no_na_df['seed'] = no_na_df['seed'].astype('Int64')
no_na_df['venture'] = no_na_df['venture'].astype('Int64')
no_na_df['angel'] = no_na_df['angel'].astype('Int64')
no_na_df['private_equity'] = no_na_df['private_equity'].astype('Int64')
no_na_df['debt_financing'] = no_na_df['debt_financing'].astype('Int64')
no_na_df['round_A'] = no_na_df['round_A'].astype('Int64')
no_na_df['round_B'] = no_na_df['round_B'].astype('Int64')
no_na_df['round_C'] = no_na_df['round_C'].astype('Int64')
no_na_df['funding_rounds'] = no_na_df['funding_rounds'].astype('Int64')
```

```python
no_na_df.head(10)
```

```python
no_na_df.seed.replace(0, np.nan, inplace=True)
no_na_df.venture.replace(0, np.nan, inplace=True)
no_na_df.angel.replace(0, np.nan, inplace=True)
no_na_df.private_equity.replace(0, np.nan, inplace=True)
no_na_df.debt_financing.replace(0, np.nan, inplace=True)
no_na_df.round_A.replace(0, np.nan, inplace=True)
no_na_df.round_B.replace(0, np.nan, inplace=True)
no_na_df.round_C.replace(0, np.nan, inplace=True)
```
